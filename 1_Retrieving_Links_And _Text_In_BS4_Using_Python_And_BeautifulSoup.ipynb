```python
import bs4
import requests
```

```python
url = "https://www.w3schools.com/python/default.asp"
```

# Get Whole Html Any Web Site

```python
data = requests.get(url)
soup = bs4.BeautifulSoup(data.text, 'html.parser')

print(soup.prettify())
```

# When Using soup.find They Print 1st Paragraph

```python
# When Using soup.find They Print 1st Paragraph
for para in soup.find('p'):
    print(para)
```

# When Using soup.find_all They Print all Paragraph With href

```python
# When Using soup.find_all They Print all Paragraph
for para in soup.find_all('p'):
    print(para)
```

# When Using soup.find_all They Print all Paragraph

```python
# When Using soup.find_all They Print all Paragraph
for para in soup.find_all('p'):
    print(para.text)
```

# Fetch All Links From Given Url

```python
# Fetch All Links From Given Url
for links in soup.find_all('a'):
    link = links.get('href')
    print(link)
```

```python
# Fetch All Links From Given Url
for links in soup.find_all('a'):
    link = links.get('href')
    if link[0:3] == "../" and link != "#":
        print("https://www.w3schools.com" + link[2:len(link)])
    elif link[0] == "/" and link != "#":
        print("https://www.w3schools.com" + link[2:len(link)])
    elif link != "#":
        print(link)
```
